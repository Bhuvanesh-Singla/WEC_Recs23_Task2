```python
import nltk
nltk.download ('punkt')
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
import gensim
from gensim.models import Word2Vec
```

```python
sample = open('alice.txt')
```

```python
s = sample.read()
```

```python
f = s.replace('\n', ' ')
```

```python
import re
def clean(text):
# Removes all special characters and numericals leaving the alphabets
    text = re.sub('[^A-Za-z]+', ' ', text)
    return text

text = clean(f)
```

```python
data = []
```

```python
for sent in sent_tokenize(text):
  temp  = []

  for word in word_tokenize(sent):
    temp.append(word.lower())

  data.append(temp)
```

```python
data
```

```python
model = gensim.models.Word2Vec(data, min_count = 1,
                              vector_size = 100, window = 5, sg = 1)
```

```python
print("Cosine similarity between 'alice' " +
               "and 'wonderland' - SkipGram : ",
    model.wv.similarity('alice', 'wonderland'))

print("Cosine similarity between 'alice' " +
                 "and 'machines' - SkipGram : ",
      model.wv.similarity('alice', 'machines'))
```

```python
word_embeddings = {word: model.wv[word] for word in model.wv.index_to_key}
```

```python
import csv

with open('word_embeddings_w2v_skipgram.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["word"] + [f"embedding_{i}" for i in range(model.vector_size)])  # Header
    for word, embedding in word_embeddings.items():
        writer.writerow([word] + list(embedding))
```
